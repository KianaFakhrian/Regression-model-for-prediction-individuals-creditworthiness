```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('CreditPrediction.csv')
```

```python
df = df.replace('', pd.NA)
df = df.replace('Unknown', pd.NA)
```

```python
missing_percent = (df.isna().sum() / len(df)) * 100
print(missing_percent)
```

# fill the categorical missing rows with mode

```python
cat_cols = ['Gender', 'Education_Level', 'Marital_Status', 'Card_Category']

for col in cat_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
df
```

```python
missing_percent = (df.isna().sum() / len(df)) * 100
```

```python
missing_percent
```

# erase missing values of Months_on_book and Total_Relationship_Count columns

```python
df = df.dropna(subset=['Months_on_book', 'Total_Relationship_Count'])
missing_percent = (df.isna().sum() / len(df)) * 100
missing_percent
```
